```python
# 3rd
class data:
    def __init__(self):
        self.airport = [
            ("Bangkok Suvarnabhumi(BKK)", "BKK"),
            ("Chiang Mai(CNX)", "CNX"),
            ("Phuket(HKT)", "HKT"),
            ("Khon Kaen(KKC)", "KKC"),
            ("Nakhon Si Thammarat(NST)", "NST"),
        ]
        self.airport2 = [
            ("Chiang Mai(CNX)", "CNX"),
            ("Phuket(HKT)", "HKT"),
            ("Khon Kaen(KKC)", "KKC"),
            ("Nakhon Si Thammarat(NST)", "NST"),
        ]
        self.table_head = [ "From", "To", "Flight No.", "Departure", "Arrival", "Time Interval", "Economy", "Business", "First-Class"]
        self.name_airport = [ "Bangkok Suvarnabhumi", "Chiang Mai", "Phuket", "Khon Kaen", "Nakhon Si Thammarat"]
        self.code_airport = ["BKK", "CNX", "HKT", "KKC", "NST"]
        self.c_airport = ["B", "C", "H", "K", "N"]
        self.path = ["BC", "BH", "BK", "BN", "CB", "CH", "CK", "CN", "HB", "HC", "HK", "HN", "KB", "KC", "KH", "KN", "NB",  "NC", "NH", "NK"]
        self.column = [ "From", "To", "Departure", "Arrival", "Economy", "Business", "First-Class", "Time Interval"]
        self.passenger = [
            "Name", "Surname", "Title", "ID", "Birthdate", "Seat", "Cont_name", "Cont_surname", "Cont_phone", "Cont_email"]

    def generate_date_str(self, date_str):
        year = str(date_str).split("-")[0]
        month = str(date_str).split("-")[1]
        date = str(date_str).split("-")[2]
        year = int(year)
        month = int(month)
        date = int(date)
        month_n = datetime.datetime(year, month, date).strftime("%B")
        day = datetime.datetime(year, month, date).strftime("%a")
        self.date_comp = (
            str(day) + " " + str(date) + " " + str(month_n) + " " + str(year)
        )
        return self.date_comp

    def decode_date(self, code_str):
        date = int(code_str[0]) + ((ord(code_str[1]) - ord("A")) * 10)
        month = ord(code_str[2]) - ord("D")
        year = 2020 + int(code_str[3])
        full = datetime.datetime(year, month, date)
        new_m = full.strftime("%B")
        self.decode = str(date) + " " + str(new_m) + " " + str(year)
        return self.decode

    def time_format(self, interval):
        hr = interval.split(":")[0] + " h "
        min = interval.split(":")[1] + " m)"
        self.tim = " (" + hr + min
        return self.tim

    def generate_flight_code(self, dep, arr, date, month, year):
        date = int(date)
        month = int(month)
        year = int(year)
        date_a = date // 10
        d = chr(date_a + ord("A"))
        date = date % 10
        d = str(date) + d
        m = chr(month + 68)
        year = year % 10
        y = str(year)
        self.code = "SP" + d + m + y + dep + arr
        return self.code


def ticket_re():
    # SP2BO1BCP
    code = search._reserv.value
    dat = data()
    date_str = code[2] + code[3] + code[4] + code[5]
    date = dat.decode_date(date_str)
    dep = dat.name_airport[dat.c_airport.index(code[6])]
    des = dat.name_airport[dat.c_airport.index(code[7])]
    dep_code = dat.code_airport[dat.c_airport.index(code[6])]
    des_code = dat.code_airport[dat.c_airport.index(code[7])]
    period = ord(code[8]) - ord("P")
    pr = csv.df
    p = 0
    id = 0
    while p < 80:
        if pr.loc[p, "From"] == dep_code and pr.loc[p, "To"] == des_code:
            id = p
            break
        p = p + 4
    time_dep = pr.loc[id, "Departure"]
    time_des = pr.loc[id, "Arrival"]

    row = search.name.row
    cor = "F" + str(row)
    cor2 = "C" + str(row)
    worksheet = csv.db.worksheet(code)
    seat = worksheet.get(cor).first()
    prefix = worksheet.get(cor2).first()
    ticket_re.label = widgets.Label(value="Your ticket")
    ticket_re.flight1 = widgets.Label(value="From " + dep)
    ticket_re.flight2 = widgets.Label(value="To " + des)
    ticket_re.flight_no = widgets.Label(value=code)
    ticket_re.fly_time = widgets.Label(value=time_dep + " - " + time_des)
    ticket_re.name = widgets.Label(
        value=prefix + "   " + search._name.value + "     " + search._surname.value
    )
    ticket_re.seat_n = widgets.Label(value=seat)

    box_layout = Layout(
        display="flex", flex_flow="row", align_items="center", width="30%"
    )
    butt = [
        widgets.Button(description="Back"),
        widgets.Button(description="Cancel Booking"),
    ]
    ticket_re.button = Box(children=butt, layout=box_layout)

    toggle_ticket_re_display(True)

    def on_button_clicked_back(b):
        toggle_ticket_re_display(False)
        with Welcome_to_Supatipunno_Airline.t.output_to(1, select=False):
            search()

    def on_button_clicked_cancel(b):
        passen = int(worksheet.get("A1").first())
        if passen == 1:
            worksheet.delete_rows(row)
        elif row == passen + 1:
            worksheet.delete_rows(row)
        else:
            values_list = worksheet.row_values(1 + passen)
            worksheet.delete_rows(passen + 1)
            list2 = [values_list]
            cor = "A" + str(row) + ":J" + str(row)
            worksheet.update(cor, list2)
        worksheet.update("A1", passen - 1)

        toggle_ticket_re_display(False)
        with Welcome_to_Supatipunno_Airline.t.output_to(1, select=False):
            search()

    ticket_re.button.children[0].on_click(on_button_clicked_back)
    ticket_re.button.children[1].on_click(on_button_clicked_cancel)


def search():
    dat = data()
    search.label = widgets.Label(value="Ticket Searching System")
    search._reserv = widgets.Text(
        value="", placeholder="SPXXXXXX", description="Flight No: ", disabled=False
    )
    search._name = widgets.Text(
        value="", placeholder="", description="First Name: ", disabled=False
    )
    search._surname = widgets.Text(
        value="", placeholder="", description="Surname: ", disabled=False
    )
    search._dep = widgets.Dropdown(
        options=dat.airport,
        description="Departure:",
    )
    search.button1 = widgets.Button(description="Search Reservation")

    search.error = 0

    # ticket search button clicked
    def on_button_clicked_search(b):
        if search.error == 1:
            search.not_found.close()
        df = csv.db
        valid = False
        try:
            worksheet = df.worksheet(search._reserv.value)
            valid = True
        except:
            search.error = 1
            search.not_found = widgets.Label(value="Reservation not found")
            display(search.not_found)

        if valid:
            worksheet = df.worksheet(search._reserv.value)
            search.name = worksheet.find(search._name.value)
            search.surname = worksheet.find(search._surname.value)
            if search.name != None and search.surname != None:
                if search.name.row == search.surname.row:
                    toggle_search_display(False)
                    with Welcome_to_Supatipunno_Airline.t.output_to(1, select=False):
                        ticket_re()
                else:
                    search.error = 1
                    search.not_found = widgets.Label(value="Reservation not found")
                    display(search.not_found)
            else:
                search.error = 1
                search.not_found = widgets.Label(value="Reservation not found")
                display(search.not_found)

    search.button1.on_click(on_button_clicked_search)
    toggle_search_display(True)


def ADD_PASSENGER():
    passenger = []
    crew = 0
    dat = data()
    ps = csv.flight_df
    up = csv.flight_list  # worksheet
    code = flight._flight.value
    dep = code[6] + code[7]
    flight_ava = ps.loc[0, dep]
    idx = chr(dat.path.index(dep) + ord("A"))
    if flight_ava == "0":
        idx = idx + "1:" + idx + "2"
        up.update(idx, [["1"], [code]])
        sheet_created = False
        while sheet_created == False:
            try:
                worksheet = csv.db.add_worksheet(title=code, rows="80", cols="10")
                worksheet = csv.db.worksheet(code)
                sheet_created = True
            except:
                pass
    else:
        id = dat.path.index(dep) + 1
        flight_list = up.col_values(id)
        flight_list.pop(0)
        low = 0
        high = len(flight_list)
        valid = False
        while low < high:
            mid = (low + high) // 2
            if flight_list[mid] == code:
                valid = True
                break
            elif flight_list[mid] < code:
                low = mid + 1
            else:
                high = mid

        if valid == False:  # invalid
            sheet_created = False
            while sheet_created == False:
                try:
                    worksheet = csv.db.add_worksheet(title=code, rows="80", cols="10")
                    worksheet = csv.db.worksheet(code)
                    sheet_created = True
                except:
                    pass
            worksheet.update("A1", "0")
            crew = 0
            toInsert = low
            flight_list.insert(toInsert, code)
            begin = idx + "2"
            last = idx + str(int(flight_ava) + 2)
            newidx = begin + ":" + last
            new_ava = idx + "1"
            newlist = []
            for i in range(len(flight_list)):
                y = []
                y.append(flight_list[i])
                newlist.append(y)
            up.update(newidx, newlist)
            up.update(new_ava, str(int(flight_ava) + 1))

        else:  # valid
            worksheet = csv.db.worksheet(code)
            crew = int(worksheet.get("A1").first())

    worksheet = csv.db.worksheet(code)
    seat = str(flight._lvl.value).split(" ")[0]
    for i in range(Home._passenger.value):
        temp = []
        temp.append(str(inform.name_box[i].children[1].value))
        temp.append(str(inform.name_box[i].children[2].value))
        temp.append(str(inform.name_box[i].children[0].value))
        temp.append(str(inform.etc_box[i].children[0].value))
        temp.append(str(inform.etc_box[i].children[1].value))
        temp.append(seat)
        temp.append(str(inform.cont_name.children[0].value))
        temp.append(str(inform.cont_name.children[1].value))
        temp.append(str(inform.cont_mail.children[0].value))
        temp.append(str(inform.cont_mail.children[1].value))
        passenger.append(temp)

    cor = "A" + str(2 + crew) + ":J" + str(1 + crew + Home._passenger.value)
    worksheet.update(cor, passenger)
    worksheet.update("A1", str(crew + Home._passenger.value))

    if Home._path.value == "R":
        code2 = flight._flight2.value
        dep2 = code[6] + code[7]
        flight_ava2 = ps.loc[0, dep2]
        idx2 = chr(dat.path.index(dep2) + ord("A"))
        crew2 = 0
        if flight_ava2 == "0":
            idx2 = idx2 + "1:" + idx2 + "2"
            up.update(idx2, [["1"], [code2]])
            sheet_created2 = False
            while sheet_created2 == False:
                try:
                    worksheet2 = csv.db.add_worksheet(title=code2, rows="80", cols="10")
                    worksheet2 = csv.db.worksheet(code2)
                    sheet_created2 = True
                except:
                    pass
        else:
            id2 = dat.path.index(dep2) + 1
            flight_list2 = up.col_values(id2)
            flight_list2.pop(0)

            low2 = 0
            high2 = len(flight_list2)
            valid2 = False

            while low2 < high2:
                mid2 = (low2 + high2) // 2
                if flight_list2[mid2] == code2:
                    valid2 = True
                    break
                elif flight_list2[mid2] < code2:
                    low2 = mid2 + 1
                else:
                    high2 = mid2

            if valid2 == False:  # invalid
                sheet_created2 = False

                while sheet_created2 == False:
                    try:
                        worksheet2 = csv.db.add_worksheet(
                            title=code2, rows="80", cols="10"
                        )
                        worksheet2 = csv.db.worksheet(code2)
                        sheet_created2 = True
                    except:
                        pass

                worksheet2.update("A1", "0")
                crew2 = 0
                toInsert2 = low2
                flight_list2.insert(toInsert2, code2)
                begin2 = idx2 + "2"
                last2 = idx2 + str(len(flight_list2) + 1)
                newidx2 = begin2 + ":" + last2
                new_ava2 = idx2 + "1"
                newlist2 = []
                for i in range(len(flight_list2)):
                    y2 = []
                    y2.append(flight_list2[i])
                    newlist2.append(y2)
                up.update(newidx2, newlist2)  #
                up.update(new_ava2, str(len(flight_list2)))

            else:  # valid
                worksheet2 = csv.db.worksheet(code)
                sheetPD2 = worksheet2.get_all_values()
                sheetDF2 = pd.DataFrame(sheetPD2)
                sheetDF2.columns = dat.passenger
                crew2 = int(worksheet2.get("A1").first())

        seat2 = str(flight._lvl2.value).split(" ")[0]
        for i in range(Home._passenger.value):
            passenger[i][5] = seat2

        cor2 = "A" + str(2 + crew2) + ":J" + str(1 + crew2 + Home._passenger.value)
        worksheet2.update(cor2, passenger)
        worksheet2.update("A1", str(crew2 + Home._passenger.value))


def ticket():
    ADD_PASSENGER()
    dat = data()
    ticket.code = flight._flight.value
    pr = csv.df
    p = 0
    ticket.id = 0
    while p < 80:
        if (
            pr.loc[p, "From"] == Home._Dep.value
            and pr.loc[p, "To"] == Home._destination.value
        ):
            ticket.id = p
            break
        p = p + 4
    flight_order = ord(ticket.code[8]) - ord("P")
    time_in = str(pr.loc[ticket.id + flight_order, "Time Interval"])
    time_dep = str(pr.loc[ticket.id + flight_order, "Departure"])
    time_ret = str(pr.loc[ticket.id + flight_order, "Arrival"])
    ticket.info1_1 = dat.name_airport[dat.c_airport.index(ticket.code[6])]
    ticket.info1_2 = dat.name_airport[dat.c_airport.index(ticket.code[7])]
    ticket.date1 = dat.generate_date_str(Home._Dep_date.value)
    ticket.time = time_dep + " - " + time_ret + dat.time_format(time_in)
    ticket._lvl = str(flight._lvl.value).split(" ")[2]
    ticket.price = int(Home._passenger.value) * float(ticket._lvl)
    items_layout = Layout(width="auto")
    box_layout = Layout(
        display="flex", flex_flow="column", align_items="center", width="14%"
    )

    word1 = ["Departure Flight", ticket.code]
    word2 = [ticket.info1_1, "To", ticket.info1_2]
    word3 = [ticket.date1, ticket.time]
    item1 = [widgets.Label(value=i, layout=items_layout) for i in word1]
    ticket.box1 = Box(children=item1, layout=box_layout)
    item2 = [widgets.Label(value=i, layout=items_layout) for i in word2]
    ticket.box2 = Box(children=item2, layout=box_layout)
    item3 = [widgets.Label(value=i, layout=items_layout) for i in word3]
    ticket.box3 = Box(children=item3, layout=box_layout)

    if Home._path.value == "R":
        ticket.code2 = flight._flight2.value
        p = 0
        while p < 80:
            if (
                pr.loc[p, "From"] == Home._destination.value
                and pr.loc[p, "To"] == Home._Dep.value
            ):
                flight.id = p
                break
            p = p + 4
        flight_order2 = ord(ticket.code2[8]) - ord("P")
        time_in2 = str(pr.loc[p + flight_order2, "Time Interval"])
        time_dep2 = str(pr.loc[p + flight_order2, "Departure"])
        time_ret2 = str(pr.loc[p + flight_order2, "Arrival"])
        ticket.info2 = dat.name_airport[dat.c_airport.index(ticket.code2[6])]
        ticket.info2_2 = dat.name_airport[dat.c_airport.index(ticket.code2[7])]
        ticket.date2 = dat.generate_date_str(Home._Ret_date.value)
        ticket.time2 = time_dep2 + " - " + time_ret2 + dat.time_format(time_in2)
        ticket._lvl2 = str(flight._lvl2.value).split(" ")[2]
        ticket.price2 = round(
            ticket.price + (int(Home._passenger.value) * float(ticket._lvl2)), 2
        )

        word4 = ["Return Flight", ticket.code2]
        word5 = [ticket.info2, "To", ticket.info2_2]
        word6 = [ticket.date2, ticket.time2]
        word_blank = ["      ||      ", "      ||      ", "      ||      "]

        item4 = [widgets.Label(value=i, layout=items_layout) for i in word4]
        ticket.box4 = Box(children=item4, layout=box_layout)
        item5 = [widgets.Label(value=i, layout=items_layout) for i in word5]
        ticket.box5 = Box(children=item5, layout=box_layout)
        item6 = [widgets.Label(value=i, layout=items_layout) for i in word6]
        ticket.box6 = Box(children=item6, layout=box_layout)
        item_blank = [widgets.Label(value=i, layout=items_layout) for i in word_blank]
        ticket.box_blank = Box(children=item_blank, layout=box_layout)

    if Home._path.value == "O":
        ticket.Box_flight = HBox([ticket.box1, ticket.box2, ticket.box3])
    elif Home._path.value == "R":
        ticket.Box_flight = HBox([ticket.box1, ticket.box2, ticket.box3, ticket.box_blank, ticket.box4, ticket.box5, ticket.box6])

    ticket.info_label = widgets.Label(value="Contact infomation")

    word_blank2 = ["            ", "            "]
    box_blank2 = [widgets.Label(value=i, layout=items_layout) for i in word_blank2]
    ticket.blank2 = Box(children=box_blank2, layout=box_layout)
    contact = []
    contact.append("Name: " + inform.cont_name.children[0].value)
    contact.append(inform.cont_name.children[1].value)
    contact.append("Phone: " + inform.cont_mail.children[0].value)
    contact.append("E-Mail: " + inform.cont_mail.children[1].value)
    ticket.contact_grid = [widgets.Label(value=i) for i in contact]
    ticket.table_contact = widgets.GridBox(
        ticket.contact_grid,
        layout=widgets.Layout(grid_template_columns="repeat(2, 225px)"),
    )

    b = [" ", " ", " ", " ", " ", " ", " ", " ", " ", 
         " ", " ", " ", " ", " ", " ", " ", " ", " ",
    ]
    blank = [widgets.Label(b[i]) for i in range(len(b))]
    ticket.table_blank = widgets.GridBox(
        blank, layout=widgets.Layout(grid_template_columns="repeat(3, 150px)")
    )

    ticket.pass_label = widgets.Label(value="Passenger Information")

    passenger = ["Name-lastname", "Citizen ID", "Birthdate"]

    for i in range(Home._passenger.value):
        s = ""
        s = (
            inform.name_box[i].children[0].value
            + "    "
            + inform.name_box[i].children[1].value
            + "        "
            + inform.name_box[i].children[2].value
        )
        passenger.append(s)
        passenger.append(inform.etc_box[i].children[0].value)
        passenger.append(inform.etc_box[i].children[1].value)

    items = [widgets.Label(str(i)) for i in passenger]
    ticket.passenger_table = widgets.GridBox(
        items, layout=widgets.Layout(grid_template_columns="repeat(3, 275px)")
    )

    ticket.home_butt = widgets.Button(description="Back to Home")

    def on_button_clicked_back(b):  # back button clicked
        toggle_ticket_display(False)
        with Welcome_to_Supatipunno_Airline.t.output_to(0, select=False):
            Home()

    toggle_ticket_display(True)

    ticket.home_butt.on_click(on_button_clicked_back)


def inform():
    dat = data()
    inform.code = flight._flight.value
    pr = csv.df
    p = 0
    inform.id = 0
    while p < 80:
        if (
            pr.loc[p, "From"] == Home._Dep.value
            and pr.loc[p, "To"] == Home._destination.value
        ):
            inform.id = p
            break
        p = p + 4
    flight_order = ord(inform.code[8]) - ord("P")
    time_in = str(pr.loc[inform.id + flight_order, "Time Interval"])
    time_dep = str(pr.loc[inform.id + flight_order, "Departure"])
    time_ret = str(pr.loc[inform.id + flight_order, "Arrival"])
    inform.info1 = (
        dat.name_airport[dat.c_airport.index(inform.code[6])]
        + " To "
        + dat.name_airport[dat.c_airport.index(inform.code[7])]
    )
    inform.date1 = dat.generate_date_str(Home._Dep_date.value)
    inform.time = time_dep + " - " + time_ret + dat.time_format(time_in)
    inform._lvl = str(flight._lvl.value).split(" ")[2]
    inform.price = int(Home._passenger.value) * float(inform._lvl)
    items_layout = Layout(width="auto")
    box_layout = Layout(
        display="flex", flex_flow="column", align_items="center", width="28%"
    )
    box_layout2 = Layout(
        display="flex", flex_flow="column", align_items="center", width="12%"
    )

    word1 = [inform.info1, inform.date1, inform.time]
    items1 = [widgets.Label(value=i, layout=items_layout) for i in word1]
    box1 = Box(children=items1, layout=box_layout)
    word2 = [" ", " ", " "]
    items2 = [widgets.Label(value=i, layout=items_layout) for i in word2]
    box2 = Box(children=items2, layout=box_layout)
    word3 = [" ", "Passenger: " + str(Home._passenger.value), " "]
    items3 = [widgets.Label(value=i, layout=items_layout) for i in word3]
    box3 = Box(children=items3, layout=box_layout2)
    word4 = [
        "Departure: " + inform._lvl + " THB",
        " ",
        "Total: " + str(inform.price) + " THB",
    ]
    items4 = [widgets.Label(value=i, layout=items_layout) for i in word4]
    box4 = Box(children=items4, layout=box_layout2)

    if Home._path.value == "R":
        inform.code2 = flight._flight2.value
        p = 0
        while p < 80:
            if (
                pr.loc[p, "From"] == Home._destination.value
                and pr.loc[p, "To"] == Home._Dep.value
            ):
                flight.id = p
                break
            p = p + 4
        flight_order2 = ord(inform.code2[8]) - ord("P")
        time_in2 = str(pr.loc[p + flight_order2, "Time Interval"])
        time_dep2 = str(pr.loc[p + flight_order2, "Departure"])
        time_ret2 = str(pr.loc[p + flight_order2, "Arrival"])
        inform.info2 = (
            dat.name_airport[dat.c_airport.index(inform.code2[6])]
            + " To "
            + dat.name_airport[dat.c_airport.index(inform.code2[7])]
        )
        inform.date2 = dat.generate_date_str(Home._Ret_date.value)
        inform.time2 = time_dep2 + " - " + time_ret2 + dat.time_format(time_in2)
        inform._lvl2 = str(flight._lvl2.value).split(" ")[2]
        inform.price2 = round(
            inform.price + (int(Home._passenger.value) * float(inform._lvl2)), 2
        )

        word5 = [inform.info2, inform.date2, inform.time2]
        items5 = [widgets.Label(value=i, layout=items_layout) for i in word5]
        box5 = Box(children=items5, layout=box_layout)
        word6 = [
            "Departure: " + inform._lvl + " THB",
            "Return: " + inform._lvl2 + " THB",
            "Total: " + str(inform.price2) + " THB",
        ]
        items6 = [widgets.Label(value=i, layout=items_layout) for i in word6]
        box6 = Box(children=items6, layout=box_layout2)

    b = [" ", " ", " ", " ", " ", " ", " ", " ", " ",
         " ", " ", " ", " ", " ", " ", " ", " ", " ",
    ]
    blank = [widgets.Label(b[i]) for i in range(len(b))]
    inform.table_blank = widgets.GridBox(
        blank, layout=widgets.Layout(grid_template_columns="repeat(3, 150px)")
    )
    inform.table_blank2 = widgets.GridBox(
        blank, layout=widgets.Layout(grid_template_columns="repeat(3, 150px)")
    )
    inform.contact_label = widgets.Label(value="Contact Information")
    inform._name = widgets.Text(
        value="", placeholder="", description="First Name: ", disabled=False
    )
    inform._surname = widgets.Text(
        value="", placeholder="", description="Surname: ", disabled=False
    )
    cont = ["First Name: ", "Surname: "]
    cont_item = [widgets.Text(value="", description=i) for i in cont]
    inform.cont_name = Box(children=cont_item)
    cont_item2 = [
        widgets.Text(value="", description="Phone: "),
        widgets.Text(value="", description="E-mail: "),
    ]
    inform.cont_mail = Box(children=cont_item2)
    inform.cont_box = VBox([inform.cont_name, inform.cont_mail])

    inform.passen_label = []
    inform.name_box = []
    inform.etc_box = []
    inform.passenger_box = []
    name_item = []
    etc_item = []
    for i in range(Home._passenger.value):
        inform.passen_label.append(widgets.Label(value="Passenger #" + str(i + 1)))
        name_item = []
        name_item.append(
            widgets.Dropdown(options=["Mr", "Mrs", "Ms"], description="title")
        )
        name_item.append(widgets.Text(value="", description="First name"))
        name_item.append(widgets.Text(value="", description="Surname"))
        inform.name_box.append(Box(children=name_item))
        etc_item = []
        etc_item.append(
            widgets.Text(description="Citizen ID", placeholder="xxxxxxxxxxxxx")
        )
        etc_item.append(widgets.DatePicker(disable=False, description="Birth Date"))
        inform.etc_box.append(Box(children=etc_item))
        inform.passenger_box.append(VBox([inform.name_box[i], inform.etc_box[i]]))

    inform.back = widgets.Button(description="Back")
    inform.next = widgets.Button(description="Book the ticket")
    butt_des = ["Back", "Next"]
    butt_item = [widgets.Button(description=i) for i in butt_des]
    inform.butt_box = Box(children=butt_item)
    inform.butt_dis = HBox([inform.butt_box])

    if Home._path.value == "O":
        inform.flight_table = HBox([box1, box2, box3, box4])
    elif Home._path.value == "R":
        inform.flight_table = HBox([box1, box5, box3, box6])

    inform.table_blank3 = widgets.GridBox(
        blank, layout=widgets.Layout(grid_template_columns="repeat(3, 150px)")
    )

    it_layout1 = Layout(
        display="flex", flex_flow="column", align_items="center", width="10%"
    )
    it_layout2 = Layout(
        display="flex", flex_flow="column", align_items="flex-start", width="60%"
    )

    inform.it_checkbox1 = widgets.Checkbox(
        value=False,
        description="I have check and confirm that all of the information are correct and cannot be edited after reservation.",
        layout={"width": "max-content"},
    )
    inform.it_checkbox2 = widgets.Checkbox(
        value=False,
        description="I have read and accept all the terms & conditions.",
        layout={"width": "max-content"},
    )

    toggle_inform_display(True)
    inform.error = 0
    num = Home._passenger.value

    def on_button_clicked_next(b):
        if inform.error == 1:
            inform.error_label1.close()
        if inform.error == 2:
            inform.error_label2.close()
        if inform.error == 3:
            inform.error_label3.close()
        if inform.error == 4:
            inform.error_label4.close()
        if inform.error == 5:
            inform.error_label5.close()
        if inform.error == 6:
            inform.error_label6.close()
        if inform.error == 7:
            inform.error_label7.close()

        if inform.it_checkbox1.value == False or inform.it_checkbox2.value == False:
            inform.error = 7
            inform.error_label7 = widgets.Label(
                value="You must accept all the policy agreement."
            )
            display(inform.error_label7)
        elif (inform.cont_name.children[0].value == "" or inform.cont_name.children[1].value == "" 
              or inform.cont_mail.children[0].value == "" or inform.cont_mail.children[1].value == "" ):
            inform.error = 1
            inform.error_label1 = widgets.Label(
                value="Please complete the contact information."
            )
            display(inform.error_label1)
        elif len(inform.cont_mail.children[0].value) != 10:
            inform.error = 2
            inform.error_label2 = widgets.Label(
                value="Please complete 10 digits of phone contact."
            )
            display(inform.error_label2)
        elif (any(domain in inform.cont_mail.children[1].value for domain in ["@gmail.com", "@outlook.com", "@hotmail.com", "@yahoo.com", "@msn.com"]) == False ):
            inform.error = 3
            inform.error_label3 = widgets.Label(
                value="Your email domain is not supported."
            )
            display(inform.error_label3)
        else:
            try:
                phone = int(inform.cont_mail.children[0].value)
                for i in range(num):
                    id = int(inform.etc_box[i].children[0].value)
            except:
                inform.error = 4
                inform.error_label4 = widgets.Label(
                    value="Please inform phone/citizen ID with numbers only."
                )
                display(inform.error_label4)
            comp = True
            for i in range(num):
                for j in range(1, 2):
                    if inform.name_box[i].children[j].value == "":
                        comp = False
                        break
                    if (inform.etc_box[i].children[0].value == "" or inform.etc_box[i].children[1].value == None ):
                        comp = False
                        break
                    elif len(inform.etc_box[i].children[0].value) != 13:
                        inform.error = 6
                        comp = False
                        break
            if comp:
                toggle_inform_display(False)
                with Welcome_to_Supatipunno_Airline.t.output_to(0, select=False):
                    ticket()

            else:
                if inform.error == 0:
                    inform.error = 5
                    inform.error_label5 = widgets.Label(
                        value="Please complete the passenger(s) information"
                    )
                    display(inform.error_label5)
                elif inform.error == 6:
                    inform.error_label6 = widgets.Label(
                        value="Please complete 13 digits of citizen ID."
                    )
                    display(inform.error_label6)

    def on_button_clicked_back(b):
        toggle_inform_display(False)
        with Welcome_to_Supatipunno_Airline.t.output_to(0, select=False):
            flight()

    inform.butt_box.children[0].on_click(on_button_clicked_back)
    inform.butt_box.children[1].on_click(on_button_clicked_next)


def toggle_ticket_re_display(toggle):
    if toggle == False:
        ticket_re.label.close()
        ticket_re.flight1.close()
        ticket_re.flight2.close()
        ticket_re.flight_no.close()
        ticket_re.fly_time.close()
        ticket_re.name.close()
        ticket_re.seat_n.close()
        ticket_re.button.close()
    else:
        display(ticket_re.label)
        display(ticket_re.flight1)
        display(ticket_re.flight2)
        display(ticket_re.flight_no)
        display(ticket_re.fly_time)
        display(ticket_re.name)
        display(ticket_re.seat_n)
        display(ticket_re.button)


def flight():
    from ipywidgets import widgets

    dat = data()
    flight.back = widgets.Button(description="Back")
    flight.next = widgets.Button(description="Next")
    items_box = [flight.back, flight.next]
    box_layout = Layout(
        display="flex", flex_flow="row", align_items="stretch", width="20%"
    )
    flight.box_auto = Box(children=items_box, layout=box_layout)
    dep = str(Home._Dep.value)[0]
    arr = str(Home._destination.value)[0]
    year = str(Home._Dep_date.value).split("-")[0]
    month = str(Home._Dep_date.value).split("-")[1]
    date = str(Home._Dep_date.value).split("-")[2]
    flight.code = []
    for i in range(4):
        flight.code.append(
            dat.generate_flight_code(dep, arr, date, month, year) + chr(i + 80)
        )
    temp = list(dat.table_head)
    pr = csv.df
    p = 0
    while p < 80:
        if (
            pr.loc[p, "From"] == Home._Dep.value
            and pr.loc[p, "To"] == Home._destination.value
        ):
            flight.id = p
            break
        p = p + 4
    for i in range(flight.id, flight.id + 4):
        temp.append(pr.loc[i, "From"])
        temp.append(pr.loc[i, "To"])
        temp.append(flight.code[i - flight.id])
        temp.append(pr.loc[i, "Departure"])
        temp.append(pr.loc[i, "Arrival"])
        temp.append(pr.loc[i, "Time Interval"])
        temp.append(pr.loc[i, "Economy"])
        temp.append(pr.loc[i, "Business"])
        temp.append(pr.loc[i, "First-Class"])
    temp = tuple(temp)
    items = [widgets.Label(temp[i]) for i in range(len(temp))]
    flight.table_depart = widgets.GridBox(
        items, layout=widgets.Layout(grid_template_columns="repeat(9, 150px)")
    )
    flight.code = tuple(flight.code)
    economy = "Economy - " + pr.loc[flight.id, "Economy"] + " THB"
    business = "Business - " + pr.loc[flight.id, "Business"] + " THB"
    firstclass = "First-Class - " + pr.loc[flight.id, "First-Class"] + " THB"
    flight._flight = widgets.Dropdown(options=flight.code, description="Select Flight:")
    flight._lvl = widgets.Dropdown(
        options=[economy, business, firstclass], description="Select seat:"
    )
    b = [" ", " ", " ", " ", " ", " ", " ", " ", " ", 
         " ", " ", " ", " ", " ", " ", " ", " ", " ", 
         " ", " ", " ", " ", " ", " ", " ", " ", " ",
    ]
    blank = [widgets.Label(b[i]) for i in range(len(b))]
    flight.table_blank = widgets.GridBox(
        blank, layout=widgets.Layout(grid_template_columns="repeat(3, 150px)")
    )
    # ########### return ############

    if Home._path.value == "R":
        arr = str(Home._Dep.value)[0]
        dep = str(Home._destination.value)[0]
        year = str(Home._Ret_date.value).split("-")[0]
        month = str(Home._Ret_date.value).split("-")[1]
        date = str(Home._Ret_date.value).split("-")[2]
        flight.code2 = []
        for i in range(4):
            flight.code2.append(
                dat.generate_flight_code(dep, arr, date, month, year) + chr(i + 80)
            )
        temp2 = list(dat.table_head)
        pr = csv.df
        p = 0
        while p < 80:
            if (
                pr.loc[p, "From"] == Home._destination.value
                and pr.loc[p, "To"] == Home._Dep.value
            ):
                flight.idx = p
                break
            p = p + 4
        for i in range(flight.idx, flight.idx + 4):
            temp2.append(pr.loc[i, "From"])
            temp2.append(pr.loc[i, "To"])
            temp2.append(flight.code2[i - flight.idx])
            temp2.append(pr.loc[i, "Departure"])
            temp2.append(pr.loc[i, "Arrival"])
            temp2.append(pr.loc[i, "Time Interval"])
            temp2.append(pr.loc[i, "Economy"])
            temp2.append(pr.loc[i, "Business"])
            temp2.append(pr.loc[i, "First-Class"])
        temp2 = tuple(temp2)
        items2 = [widgets.Label(temp2[i]) for i in range(len(temp2))]
        flight.table_return = widgets.GridBox(
            items2, layout=widgets.Layout(grid_template_columns="repeat(9, 150px)")
        )
        flight.code2 = tuple(flight.code2)
        economy = "Economy - " + pr.loc[flight.idx, "Economy"] + " THB"
        business = "Business - " + pr.loc[flight.idx, "Business"] + " THB"
        firstclass = "First-Class - " + pr.loc[flight.idx, "First-Class"] + " THB"

        flight._flight2 = widgets.Dropdown(
            options=flight.code2, description="Select Flight:"
        )
        flight._lvl2 = widgets.Dropdown(
            options=[economy, business, firstclass], description="Select seat:"
        )
        flight._flight2.observe(on_change_flight2, names="value")

    flight.dep_label = widgets.Label(value="Departure Flight")
    flight.ret_label = widgets.Label(value="Return Flight")

    toggle_flight_display(True)

    def on_button_clicked_back(b):  # back button clicked
        sys.stdout.write("\r" + "")
        sys.stdout.write("\033[K")
        toggle_flight_display(False)
        with Welcome_to_Supatipunno_Airline.t.output_to(0, select=False):
            Home()

    def on_button_clicked_next(b):  # next button clicked

        sys.stdout.write("\r" + "")
        sys.stdout.write("\033[K")
        toggle_flight_display(False)
        with Welcome_to_Supatipunno_Airline.t.output_to(0, select=False):
            inform()

    flight.back.on_click(on_button_clicked_back)
    flight.next.on_click(on_button_clicked_next)
    flight._flight.observe(on_change_flight, names="value")


def Home():
    from ipywidgets import widgets

    dat = data()
    Datetime = str(datetime.datetime.now())
    _datetime = list(Datetime.split())
    Home.date = _datetime[0]
    sc = list(str(Home.date).split("-"))  # current date
    yc = int(sc[0])
    mc = int(sc[1])
    dc = int(sc[2])
    time = _datetime[1]
    Home.button1 = widgets.Button(description="Search flight")
    Home._Dep = widgets.Dropdown(options=dat.airport, description="From:")
    Home._destination = widgets.Dropdown(options=dat.airport2, description="To:")
    Home._path = widgets.RadioButtons(
        options=[("One Way", "O"), ("Round Trip", "R")],
        description="Booking ",
        disabled=False,
    )
    Home._Dep_date = widgets.DatePicker(
        value=datetime.date(yc, mc, dc), description="Depart date", disabled=False
    )
    Home._Ret_date = widgets.DatePicker(
        value=datetime.date(yc, mc, dc), description="Return date", disabled=True
    )
    Home._passenger = widgets.BoundedIntText(
        value=1,
        min=1,
        max=80,
        step=1,
        description="Passenger:",
        disabled=False,
    )

    toggle_home_display(True)

    def on_button_clicked_search(b):  # Search flight button clicked
        if Home._path.value == "O" and Home._Dep_date.value is None:
            sys.stdout.write(
                "\r" + "\033[1m" + "Please select departure date" + "\033[0m"
            )
        elif Home._path.value == "R" and (
            Home._Dep_date.value is None or Home._Ret_date.value is None
        ):
            sys.stdout.write(
                "\r" + "\033[1m" + "Please select departure/return date" + "\033[0m"
            )
        else:
            sys.stdout.write("\r" + "")
            sys.stdout.write("\033[K")
            toggle_home_display(False)
            with Welcome_to_Supatipunno_Airline.t.output_to(0, select=False):
                flight()

    Home.button1.on_click(on_button_clicked_search)
    Home._Dep.observe(on_change_port, names="value")
    Home._path.observe(on_change_path, names="value")
    Home._Dep_date.observe(on_change_date, names="value")
    Home._Ret_date.observe(on_change_date, names="value")


def toggle_search_display(toggle):
    if toggle == False:
        search.label.close()
        search._reserv.close()
        search._name.close()
        search._surname.close()
        search._dep.close()
        search.button1.close()
    else:
        display(search.label)
        display(search._reserv, search._name, search._surname)
        display(search._dep)
        display(search.button1)


def toggle_ticket_display(toggle):
    if toggle == False:
        ticket.Box_flight.close()
        ticket.blank2.close()
        ticket.info_label.close()
        ticket.table_contact.close()
        ticket.table_blank.close()
        ticket.pass_label.close()
        ticket.passenger_table.close()
        ticket.home_butt.close()
    else:
        display(ticket.Box_flight)
        display(ticket.blank2)
        display(ticket.info_label)
        display(ticket.table_contact)
        display(ticket.table_blank)
        display(ticket.pass_label)
        display(ticket.passenger_table)
        display(ticket.home_butt)


def toggle_flight_display(toggle):
    if toggle == False:
        flight.table_depart.close()
        flight._flight.close()
        flight._lvl.close()
        flight.box_auto.close()
        flight.dep_label.close()
        if Home._path.value == "R":
            flight.table_blank.close()
            flight.ret_label.close()
            flight.table_return.close()
            flight._flight2.close()
            flight._lvl2.close()
    else:
        if Home._path.value == "O":
            display(flight.dep_label)
            display(flight.table_depart)
            display(flight._flight)
            display(flight._lvl)
            display(VBox([flight.box_auto]))
        elif Home._path.value == "R":
            display(flight.dep_label)
            display(flight.table_depart)
            display(flight._flight)
            display(flight._lvl)
            display(flight.table_blank)
            display(flight.ret_label)
            display(flight.table_return)
            display(flight._flight2)
            display(flight._lvl2)
            display(VBox([flight.box_auto]))


def toggle_inform_display(toggle):
    if toggle == False:
        inform.flight_table.close()
        inform.table_blank.close()
        inform.contact_label.close()
        inform.cont_box.close()
        inform.table_blank2.close()
        for i in range(Home._passenger.value):
            inform.passen_label[i].close()
            inform.passenger_box[i].close()
        inform.table_blank3.close()
        inform.it_checkbox1.close()
        inform.it_checkbox2.close()
        inform.butt_dis.close()

    else:
        display(inform.flight_table)
        display(inform.table_blank)
        display(inform.contact_label)
        display(inform.cont_box)
        display(inform.table_blank2)
        for i in range(Home._passenger.value):
            display(inform.passen_label[i])
            display(inform.passenger_box[i])
        display(inform.table_blank3)
        display(inform.it_checkbox1)
        display(inform.it_checkbox2)
        display(inform.butt_dis)


def toggle_home_display(toggle):
    if toggle == False:
        Home._path.close()
        Home._Dep.close()
        Home._destination.close()
        Home._Dep_date.close()
        Home._Ret_date.close()
        Home._passenger.close()
        Home.button1.close()
    else:
        display(Home._path)
        display(Home._Dep, Home._destination)
        display(Home._Dep_date, Home._Ret_date)
        display(Home._passenger)
        display(Home.button1)


def on_change_date(change):
    if change["type"] == "change":
        sc = list(str(Home.date).split("-"))  # current date
        yc = sc[0]
        mc = sc[1]
        dc = sc[2]
        if Home._Dep_date.value != None:
            sd = list(str(Home._Dep_date.value).split("-"))  # Departure date
            yd = sd[0]
            md = sd[1]
            dd = sd[2]
            if (
                yd < yc or yd == yc and md < mc or yd == yc and md == mc and dd < dc
            ):  # pick past date
                sys.stdout.write(
                    "\r" + "\033[1m" + "Cannot select past date" + "\033[0m"
                )
                Home.button1.disabled = True
            else:
                sys.stdout.write("\r" + "")
                sys.stdout.write("\033[K")
                Home.button1.disabled = False
        if Home._Ret_date.value != None:
            sr = list(str(Home._Ret_date.value).split("-"))  # Return date
            yr = sr[0]
            mr = sr[1]
            dr = sr[2]
            if (
                yr < yc or yr == yc and mr < mc or yr == yc and mr == mc and dr < dc
            ):  # pick past date
                sys.stdout.write(
                    "\r" + "\033[1m" + "Cannot select past date" + "\033[0m"
                )
                Home.button1.disabled = True
            else:
                sys.stdout.write("\r" + "")
                sys.stdout.write("\033[K")
                Home.button1.disabled = False

        if (
            Home._Dep_date.value != None
            and Home._Ret_date.value != None
            and Home._path.value == "R"
        ):
            if (
                yd > yr or yd == yr and md > mr or yd == yr and md == mr and dd > dr
            ):  # Return date before departure date
                sys.stdout.write(
                    "\r"
                    + "\033[1m"
                    + "Return date cannot be picked before departure date"
                    + "\033[0m"
                )
                Home.button1.disabled = True
            else:
                sys.stdout.write("\r" + "")
                sys.stdout.write("\033[K")
                Home.button1.disabled = False


def on_change_flight(change):
    if change["type"] == "change":
        fli = flight._flight.value
        fli = int(ord(fli[8]) - 80)
        idx = flight.id + fli
        price = csv.df
        economy = "Economy - " + price.loc[idx, "Economy"] + " THB"
        business = "Business - " + price.loc[idx, "Business"] + " THB"
        firstclass = "First-Class - " + price.loc[idx, "First-Class"] + " THB"
        flight._lvl.options = [economy, business, firstclass]


def on_change_flight2(change):
    if change["type"] == "change":
        fli = flight._flight2.value
        fli = int(ord(fli[8]) - 80)
        idx = flight.idx + fli
        price = csv.df
        economy = "Economy - " + price.loc[idx, "Economy"] + " THB"
        business = "Business - " + price.loc[idx, "Business"] + " THB"
        firstclass = "First-Class - " + price.loc[idx, "First-Class"] + " THB"
        flight._lvl2.options = [economy, business, firstclass]


def on_change_port(change):
    if change["type"] == "change":
        dat = data()
        temp = dat.airport
        for i in range(len(temp)):
            if temp[i][1] == change["new"]:
                idx = i
                break
        del temp[idx]
        Home._destination.index = None
        Home._destination.options = temp


def on_change_path(change):
    if change["type"] == "change":
        if change["new"] == "R":
            sys.stdout.write("\r" + "")
            sys.stdout.write("\033[K")
            Home._Ret_date.disabled = False
        elif change["new"] == "O":
            sys.stdout.write("\r" + "")
            sys.stdout.write("\033[K")
            Home._Ret_date.disabled = True


def csv():
    dat = data()
    # flight price
    wb = gc.open_by_key("1_H4p2M4aEeCR-O1FOtdHTBJs64t6wloTs3V6d8qz5tI")
    csv.sheet = wb.worksheet("Flight")
    pr = csv.sheet.get_all_values()
    csv.df = pd.DataFrame(pr)
    csv.df.columns = dat.column

    # flight Database
    csv.db = gc.open_by_key("1z6YKrstF-dQ8uH4ixG33gda9XIV04g3uVFxWQhkBBX8")
    csv.flight_list = csv.db.worksheet("Flight_List")
    flight_PD = csv.flight_list.get_all_values()
    csv.flight_df = pd.DataFrame(flight_PD)
    csv.flight_df.columns = dat.path


def Welcome_to_Supatipunno_Airline():
    from google.colab import widgets
    from google.colab import output

    csv()
    Welcome_to_Supatipunno_Airline.t = widgets.TabBar(["Home", "My Booking"])
    with Welcome_to_Supatipunno_Airline.t.output_to(0, select=False):
        Home()
    with Welcome_to_Supatipunno_Airline.t.output_to(1, select=False):
        search()
```
